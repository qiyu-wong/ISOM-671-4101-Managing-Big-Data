```python
############# In this file we join tables and standardize numeric columns
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

###### joining dataset
# open all the dataset
data_dir = "C:\\Users\\dongh\\Downloads\\MBD project\\"
sex_ratio = pd.read_csv(data_dir + "sexratio.csv")
lockdown = pd.read_csv(data_dir + "lockdown.csv")
new_age = pd.read_csv(data_dir + "new_age.csv")
perform_healths = pd.read_csv(data_dir + "efficiency of health systems.csv")
new_metrics = pd.read_csv(data_dir + "new_metrics.csv")
new_pop = pd.read_csv(data_dir + "new_pop.csv")
new_tourists = pd.read_csv(data_dir + "new_tourists.csv")
central = pd.read_csv(data_dir + "centralities.csv")
#unemp = pd.read_csv(data_dir + "unemp.csv")
```

```python
# join tables one by one
df = pd.merge(sex_ratio,new_age)
df1 = pd.merge(df, perform_healths)
df2 = pd.merge(df1, new_tourists)
df3 = pd.merge(df2, lockdown)
df4 = pd.merge(df3, new_metrics)
df5 = pd.merge(df4, new_pop)
dfx = pd.merge(df5, central)
dfx
```

```python
# save the big table as csv file
dfx.to_csv("C:\\Users\\dongh\\Downloads\\MBD project\\merged4.csv")
```

```python
####### Standardize numeric columns
# read in the dataset and check data types
df = pd.read_csv("C:\\Users\\dongh\\Downloads\\MBD project\\merged4.csv")
df.dtypes
```

```python
# convert to zscore for column sex ratio
from scipy import stats
SexRatio = df['Sex Ratio']
SexRatio = np.array(SexRatio)
df['Sex Ratio'] = stats.zscore(SexRatio)
print(df)
```

```python
# standardize each of the column
MedianAge = df['MedianAge']
MedianAge = np.array(MedianAge)
df['MedianAge'] = stats.zscore(MedianAge)

c3 = df['efficiency of health systems']
c3 = np.array(c3)
df['efficiency of health systems'] = stats.zscore(c3)

df['2019 Visitation'].dtype
c5 = df['2019 Visitation']
c5 = np.array(c5)
df['2019 Visitation'] = stats.zscore(c5)

c6 = df['Gini']
c6 = np.array(c6)
df['Gini'] = stats.zscore(c6)

c7 = df['ICU Beds']
c7 = np.array(c7)
df['ICU Beds'] = stats.zscore(c7)

c8 = df['GDP']
c8 = np.array(c8)
df['GDP'] = stats.zscore(c8)

c9 = df['Hospitals']
c9 = np.array(c9)
df['Hospitals'] = stats.zscore(c9)

c10 = df['Life Expectancy']
c10 = np.array(c10)
df['Life Expectancy'] = stats.zscore(c10)

c11 = df['Population, 2019']
c11 = np.array(c11)
df['Population, 2019'] = stats.zscore(c11)

c12 = df['pop per km2']
c12 = np.array(c12)
df['pop per km2'] = stats.zscore(c12)

c12 = df['outDegree']
c12 = np.array(c12)
df['outDegree'] = stats.zscore(c12)

c12 = df['inDegree']
c12 = np.array(c12)
df['inDegree'] = stats.zscore(c12)

c12 = df['pagerank']
c12 = np.array(c12)
df['pagerank'] = stats.zscore(c12)
```
